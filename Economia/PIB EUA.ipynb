```r
library(fredr)
library(ggplot2)
fredr_set_key("a1e53d11bf4cb319624da39b894a79e6")
un <- fredr("UNRATE", observation_start = as.Date("1949-01-01"), frequency = "q", units = "chg")
gdp <- fredr("GDP", observation_start = as.Date("1949-01-01"), units = "pch",
             observation_end = as.Date("2020-01-01"))
un <- un[-nrow(un),]
un1 <- fredr("UNRATE", observation_start = as.Date("1949-01-01"), frequency = "q",
             observation_end = as.Date("2020-01-01"))
regressao <- lm(un$value ~ gdp$value)
nrate <- fredr("NROU", frequency = "q", observation_start = as.Date("1949-01-01"),
               observation_end = as.Date("2020-01-01"))
gdp1 <- fredr("GDP", observation_start = as.Date("1949-01-01"))
beta <- 1.4
out <- -beta*(un1$value-nrate$value)
p_gdp <- fredr("NGDPPOT", observation_start = as.Date("1949-01-01"), frequency = "q",
               observation_end = as.Date("2020-01-01"))
pot <- ((gdp1$value-p_gdp$value)/p_gdp$value)*100
out <- data.frame(Data = gdp$date,Gap = out, Real_Gap = pot)
ggplot(out, aes(Data))+
  geom_line(aes(y = Gap, colour = "Gap"))+
  geom_line(aes(y = Real_Gap, colour = "Real Gap"))
varout <- diff(out$Gap,1)
varpot <- diff(out$Real_Gap,1)
var_tot <- data.frame(Data = gdp$date[2:length(gdp$value)], Var_Gap = varout, Var_Real = varpot)
ggplot(var_tot, aes(x = Data))+
  geom_line(aes(y = Var_Gap, col = "Var Gap"))+
  geom_line(aes(y = Var_Real, col = "Var Real"))
```
