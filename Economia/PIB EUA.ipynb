```r
library(ggplot2)
library(ggthemes)
library(fredr)
fredr_set_key("a1e53d11bf4cb319624da39b894a79e6")
gdp_us <- fredr(series_id = "GDP", observation_start = as.Date("1947-01-01"))
data <- seq(as.Date("1948-01-01"), as.Date("2020-01-01"), "3 months")
variacao <- (diff(gdp_us$value, 4)/gdp_us$value[5:293])*100
variacao <- data.frame(Data = data, Valor = variacao)
ggplot(variacao, aes(x = data))+
  geom_line(aes(y = Valor), colour = "blue")+
  labs(x = '', y = "Variação %", title = "Variação PIB EUA em 12 meses",
       caption = "Fonte:FRED")+
  theme_economist()
```
