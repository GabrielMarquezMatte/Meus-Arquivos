```r
library(rbcb)
library(ggplot2)
expectativas_2020 <- get_annual_market_expectations("IPCA", start_date = "2020-01-01")
sequ1 <- seq(1,1020,10)
expectativas_2020 <- expectativas_2020[sequ1,]
sequ2 <- seq(97,102,1)
expectativas_2020 <- expectativas_2020[-sequ2,]
expectativas_2020_2019 <- get_annual_market_expectations("IPCA",
                                                         start_date = "2019-01-01",
                                                         end_date = "2019-12-31")
seq1_2019 <- seq(2,2530,10)
expectativas_2020_2019 <- expectativas_2020_2019[seq1_2019,]
expectativa_2020_tot <- rbind(expectativas_2020, expectativas_2020_2019)
ggplot(expectativa_2020_tot, aes(x = date, y = median))+
  geom_line(colour = "red")+
  labs(x = "Data", y = "Valor",
       title = "Expectativa anual IPCA-2020", caption = "Fonte: BCB",
       subtitle = "Mediana, mínimos e máximos das previsões")+
  geom_line(aes(y = min), colour = "blue")+
  geom_line(aes(y = max), colour = "blue")
```
