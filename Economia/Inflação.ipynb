```r
library(rbcb)
library(forecast)
library(ggplot2)
```

```r
#Autocorrelação da inflação
inf <- rbcb::get_series(c(Inflacao = 433), as = "ts", start_date = "2000-01-01")
ggAcf(inf, lag.max = 12, type = "correlation")+
  labs(title = "Correlograma inflação mensal", subtitle = "De 01/01/2001 à 01/05/2020",
       caption = "Fonte:BCB e elaboração própria")
```

```r
#Histograma dos dados
inf1 <- get_series(c(inf = 433), start_date = "2001-01-01")
ggplot(inf1, aes(x = inf))+
  geom_histogram(bins = 50, fill = "blue")+
  geom_vline(xintercept = mean(inf1$inf))+
  geom_vline(xintercept = mean(inf1$inf)+2*sd(inf1$inf), lty = 2, col = "red")+
  geom_vline(xintercept = mean(inf1$inf)-2*sd(inf1$inf), lty = 2, col = "red")+
  labs(x = "%", y = "Frequência",
       title = "Histograma Inflação Mensal",
       subtitle = paste0("Média=", round(mean(inf1$inf),2),
                         "% e Desvio=",round(sd(inf1$inf),2), "%"))
```
