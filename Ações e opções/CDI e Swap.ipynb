```r
library(ggplot2)
library(BETS)
library(dplyr)
library(ggthemes)
```

```r
swap  <- BETS::BETSget(7806)
cdi <- BETSget(4389, from = "1999-08-20", to = "2019-09-30")
tot <- left_join(cdi, swap, by = c("date" = "date"))
```

```r
tot_b <- tot %>%
  filter(date > as.Date("2003-05-01"))
```

```r
ggplot(tot_b, aes(x = date))+
  geom_line(aes(y = value.x, col = "CDI"))+
  geom_line(aes(y = value.y, col = "Taxa Swap"))+
  theme_economist()+
  scale_colour_manual("", values = c("CDI" = "red",
                                     "Taxa Swap" = "blue"))+
  labs(x = "", y = "%",
       title = "CDI e Taxa Swap de referÃªncia",
       subtitle = "Anualizado")
```
