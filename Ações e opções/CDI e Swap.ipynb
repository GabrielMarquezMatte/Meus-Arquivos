```r
options(warn = -1)
library(ggplot2)
library(rbcb)
library(dplyr)
library(ggthemes)
```

```r
swap  <- get_series(c(swap = 7806))
cdi <- get_series(c(cdi = 4389))
tot <- left_join(cdi, swap, by = c("date" = "date"))
```

```r
tot_b <- tot %>%
  filter(date > as.Date("2000-05-01"))
```

```r
ggplot(tot_b, aes(x = date))+
  geom_line(aes(y = cdi, col = "CDI"))+
  geom_line(aes(y = swap, col = "Taxa Swap"))+
  theme_economist()+
  scale_colour_manual("", values = c("CDI" = "red",
                                     "Taxa Swap" = "blue"))+
  labs(x = "", y = "%",
       title = "CDI e Taxa Swap de referÃªncia",
       subtitle = "Anualizado")
```
