```r
suppressPackageStartupMessages(library(tidyverse))
equacao <- function(r,l = 100){
  x <- matrix(nrow = l)
  x[1] <- runif(1)
  for(i in 2:nrow(x)){
    x[i] <- r*x[i-1]*(1-x[i-1])
  }
  return(x)
}
```

```r
rates <- seq(0.5,4,l = 100000)
x <- sapply(rates, equacao, l = 300)
endcap <- round(nrow(x)*0.9)
maximos <- unique(x[endcap,])
data <- data.frame(rate = rates,max = maximos)
```

```r
ggplot(data, aes(x = rate, y = max))+
  geom_point(pch = ".")+
  labs(x = "", ylab = "", title = "Logistic Map",
       caption = "Elaboração: Gabriel Matte")
```
