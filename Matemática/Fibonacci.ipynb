```r
pacotes <- c("ggplot2", "ggthemes")
for(i in pacotes){
    library(i, character.only = T)
}
```

```r
n <- 20
f <- numeric(n)
f[1] <- 0
f[2] <- 1
for(i in 3:n){
  f[i] <- f[i-1]+f[i-2]
}
f <- data.frame(Indice = 1:n, Fibo = f)
```

```r
p <- ggplot(f, aes(Indice, Fibo))+
  geom_line(col = "blue")+
  labs(x = "Índice", y = "Fibonacci", title = "Sequência de Fibonacci")
```
