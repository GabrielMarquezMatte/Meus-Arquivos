```r
cos_f <- function(x = 0, l = 5){
  repe <- c(1,0,-1,0)
  repe <- rep(repe,l)
  n <- 0:(length(repe)-1)
  sum(repe*x^n/factorial(n), na.rm = T)
}
cos_f <- Vectorize(cos_f)
```

```r
library(latex2exp)
x <- seq(-30,30,l = 1000)
coo <- cos_f(x, l = 90)
coss <- cos(x)
expre1 <- TeX("$\\sum_{i = 0}{\\infty}\\frac{d^nf}{dx^n}*\\frac{x^n}{n!}$")
plot(x,coo, type = "l", ylab = "y", main = "Cosseno pela série de Taylor")
lines(x,coss, col = "red")
legend("bottom", legend = c(expre1, "cos(x)"), col = c("black","red"),
       lty = 1, bty = "n")
```

```r
erro <- coo-coss
plot(x, erro, type = "l", xlab = "x", ylab = "y", main = "Erro da série")
```

```r
e <- function(x = 0, l = 5){
  repe <- exp(0)
  repe <- rep(repe,l)
  n <- 0:(length(repe)-1)
  sum(repe*x^n/factorial(n))
}
e <- Vectorize(e)
```
